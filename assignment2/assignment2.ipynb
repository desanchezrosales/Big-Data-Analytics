## Getting the data
We use the following code to get the data

```python
import pandas as pd
fileName_1 = "/fs/scratch/PAS1585/wine/winequality-white.csv"
fileName_2 = "/fs/scratch/PAS1585/wine/winequality-red.csv"
data_1=pd.read_csv(fileName_1,sep=';')
data_2=pd.read_csv(fileName_2,sep=';')
```

## Simple column info

---

Now loop over the column names, and get the mean and std deviation of each column

```python
def mean_std_df(df):
    names = df.columns.values
    means = []
    for i in names:
        means.append((i, df[i].mean(),df[i].std()))
    return means
```

```python
mean_std_df(data_1)
```

```python
#RED WINE
mean_std_df(data_2)
```

## The correlation matrix
Pandas has a nice feature for graphically displaying the correlation matrix of a pandas dataframe:

```python
mid = data_1['quality']
data_1.drop(labels=['quality'], axis=1,inplace = True)
data_1.insert(0, 'quality', mid)

corr1 = data_1.corr()
corr1.style.background_gradient().set_precision(3)
```

```python
#RED WINE
mid2 = data_2['quality']
data_2.drop(labels=['quality'], axis=1,inplace = True)
data_2.insert(0, 'quality', mid2)

corr2 = data_2.corr()
corr2.style.background_gradient().set_precision(3)
```

## Overlaying histograms


---

Now lets make a simple overlaid histogram.  We will make two dataframes: one for wines with quality above the mean, and one for wines with quality below the mean.    Then we can easily plot any feature from those two dataframes on the same figure.   Let's do "xxxxx" first:

```python
quality_above_mean = data_1.loc[(data_1['quality'] > data_1['quality'].mean())]
quality_below_mean = data_1.loc[(data_1['quality'] < data_1['quality'].mean())]

plt.hist(quality_above_mean['alcohol'], label ='alcohol above mean')
plt.hist(quality_below_mean['alcohol'], label = 'alcohol below mean')
plt.legend()
plt.show()
```

```python
#RED WINE
quality_above_mean = data_2.loc[(data_1['quality'] > data_2['quality'].mean())]
quality_below_mean = data_2.loc[(data_1['quality'] < data_2['quality'].mean())]

plt.hist(quality_above_mean['alcohol'], label ='alcohol above mean')
plt.hist(quality_below_mean['alcohol'], label = 'alcohol below mean')
plt.legend()
plt.show()
```
