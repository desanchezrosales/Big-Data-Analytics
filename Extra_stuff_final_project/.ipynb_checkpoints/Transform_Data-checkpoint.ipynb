```python
from PIL import Image
import glob 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.utils import shuffle


fname = '/fs/scratch/PAS1585/sanchezrosales1/train_labels/train_labels.csv'

df = pd.read_csv(fname)
```

```python
import time

start = time.time()

short = True

if short:
    pixels = []
    label = []
    for i in range(50000):
        pixels.append(np.asarray(Image.open( '/fs/scratch/PAS1585/sanchezrosales1/train/' + df['id'][i] + '.tif' )))
        label.append(df['label'][i])

else:
    pixels = []
    for i in range(len(df)):
        pixels.append(np.array(Image.open( '/fs/scratch/PAS1585/sanchezrosales1/train/' + df['id'][i] + '.tif' )))
        label.append(df['label'][i])
        
end = time.time()

print(end - start)
```

```python
fig=plt.figure(figsize=(96, 96))
columns = 5
rows = 1
for i in range(1, columns*rows +1):
    img = pixels[i-1]
    img = img.reshape(96,96,3)
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```

```python
a = np.append(pixels[0][0],pixels[0][1],axis =0)
plt.imshow(a.reshape(2,96,3))
```

```python
start = time.time()

import h5py
hf = h5py.File('/fs/scratch/PAS1585/sanchezrosales1/data.h5', 'w')
hf.create_dataset('data', data=pixels)
hf.close()

end = time.time()
print(end - start)
```
